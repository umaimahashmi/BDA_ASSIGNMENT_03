```python
import re
import dask.dataframe as dd
def extract_price(price_string):
    match = re.search(r'(\d+\.\d+)_+(\d+\.\d+)', str(price_string)) 
    if match:
        start_price = float(match.group(1))
        end_price = float(match.group(2))
        
        return (start_price + end_price) / 2  #mean of start price and end price
    else:

        match_single = re.search(r'\d+\.\d+', str(price_string))
        if match_single:
            return float(match_single.group())
        else:
            return None
# removing rows with missing values(NaN)
df = df.dropna()
df['price'] = df['price'].astype(str)
df['price'] = df['price'].apply(extract_price, meta=('price', 'float'))
```

```python
import re
import pandas as pd

def clean_text(text):
    # Check if the input is missing or null
    if pd.isna(text):
        return ""
    
    # Removing HTML tags
    text = re.sub(r'<[^>]+>', '', str(text))
    
    # Remove special characters and punctuation
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    # removing extra space
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.lower()
    return text

```
